```python
import tkinter as tk
from tkinter import messagebox
import math

class Calculator:
    def __init__(self, root):
        self.root = root
        self.root.title("Calculator")
        self.expression = ""

        # Create the display for the calculator
        self.display = tk.Entry(root, font=("Arial", 20), borderwidth=2, relief="solid", justify="right", bg="lightgray")
        self.display.grid(row=0, column=0, columnspan=5, ipadx=8, ipady=8, pady=10)

        # Define button layout
        buttons = [
            ('7', 1, 0, 1), ('8', 1, 1, 1), ('9', 1, 2, 1), ('/', 1, 3, 1), ('(', 1, 4, 1),
            ('4', 2, 0, 1), ('5', 2, 1, 1), ('6', 2, 2, 1), ('*', 2, 3, 1), (')', 2, 4, 1),
            ('1', 3, 0, 1), ('2', 3, 1, 1), ('3', 3, 2, 1), ('-', 3, 3, 1), ('^', 3, 4, 1),
            ('0', 4, 0, 1), ('C', 4, 1, 1), ('=', 4, 2, 1), ('+', 4, 3, 1), ('√', 4, 4, 1),
            ('Back', 5, 0, 5)  # Backspace button spanning 5 columns
        ]

        # Create buttons
        for (text, row, col, colspan) in buttons:
            button = tk.Button(root, text=text, font=("Arial", 20), borderwidth=1, relief="solid", 
                               bg="lightblue", command=lambda t=text: self.on_button_click(t))
            button.grid(row=row, column=col, columnspan=colspan, ipadx=10, ipady=10, padx=5, pady=5)

    def on_button_click(self, button_text):
        if button_text == "C":
            self.expression = ""
            self.display.delete(0, tk.END)
        elif button_text == "=":
            try:
                # Replace ^ with ** for power operation and √ with math.sqrt for square root
                expression = self.expression.replace('^', '**').replace('√', 'math.sqrt')
                result = str(eval(expression))
                self.display.delete(0, tk.END)
                self.display.insert(tk.END, result)
                self.expression = result
            except Exception as e:
                messagebox.showerror("Error", f"Invalid input: {e}")
                self.expression = ""
                self.display.delete(0, tk.END)
        elif button_text == "√":
            self.expression += "√("
            self.display.delete(0, tk.END)
            self.display.insert(tk.END, self.expression)
        elif button_text == "Back":
            self.expression = self.expression[:-1]
            self.display.delete(0, tk.END)
            self.display.insert(tk.END, self.expression)
        else:
            self.expression += button_text
            self.display.delete(0, tk.END)
            self.display.insert(tk.END, self.expression)

if __name__ == "__main__":
    root = tk.Tk()
    calculator = Calculator(root)
    root.mainloop()
```
